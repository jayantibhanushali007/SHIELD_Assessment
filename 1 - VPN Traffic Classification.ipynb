__My Understanding of the Problem Statement:__
- For SHIELD, One key problem is to address `Early Fraud Detection`. A common sign of a fraudulent device is the use of a VPN to hide its identity and avoid being traced.
- Hence the problem of 'accurately predicting if a certain app is a VPN app' is an important feature to determine a potential threat (fraud device).


__Use of Public Dataset:__
- There were multiple labelled dataset available for VPN/ Non-VPN classification such as ISCX VPN-nonVPN Dataset (2016), MIT Lincoln VPN-nonVPN Network Dataset.
- Based on the documentation available and the limited time for this assessment, I decided to go ahead with `ISCX VPN-nonVPN Dataset (2016)` since I could find good documentation about the dataset, some research papers using this dataset for similar objective and also since the official datasets were available in .ARFF format which can be read directly as Python DataFrame using Pandas.
- Dataset link: http://205.174.165.80/CICDataset/ISCX-VPN-NonVPN-2016/Dataset/CSVs/

__About the Dataset:__
- The ISCX-VPN2016 dataset was used to develop a model in this study. The University of New Brunswick (UNB) made this dataset available for public use. It was collected by capturing experimental network traffic on a real-time basis in a customized lab. This data set takes into account seven different types of traffic: web browsing, email, chat, streaming, file transfer, VoIP, and P2P. The traffic was captured with Wireshark and TcpDump. 
- The final processed features dataset was then fed into two analysis scenarios: Scenario A was used to detect if the traffic is VPN or not and then tried to build classifiers to distinguish the type of encrypted traffic over VPN network connectivity. Scenario B used the labeled dataset, i.e., VPN and non-VPN instances, to form the second labeled dataset, in which the applications used to conduct the network-generated traffic were used as another variation of the instancesâ€™ labeling.
- This dataset gained extra value related to the type of the features, which were either packet-based or time-based. Packet-based are those features that rely more on the protocol of the networking concept conducted, such as IPs, ports, and the flow of bytes amount. By contrast, the time-based features are considered as another dimension for the analysis that does not rely only on the statistic elements of the data, but rather on the variability of the time-based features. Timeouts for the time-based features were captured according to the discrete values that were measured in seconds, including 15, 30, 60, and 120 s.
- We will be using `30s time interval dataset` since it offers a balanced compromise between the fine-grained temporal details captured in shorter intervals (e.g., 15s) and the broader aggregated insights offered by longer intervals (e.g., 120s)

## 3: Model Development on Full Data:

```python
# Define the features (X) and target (y)
X = raw_df_scaled.drop(columns=['class1', 'class1_binary'])
y = raw_df_scaled['class1_binary']

# Split the dataset into training and testing sets (70:30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Display the shapes of the resulting datasets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python


# Initialize models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    'SVM': SVC()
}

# Train and evaluate models
results = {}
for model_name, model in models.items():
    # Cross-validation
    cv_scores = cross_val_score(model, X_train, y_train, cv=5)
    print(f"{model_name} CV Accuracy: {cv_scores.mean():.4f} Â± {cv_scores.std():.4f}")
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on test data
    y_pred = model.predict(X_test)
    
    # Classification report
    print(f"\n{model_name} Classification Report:\n")
    print(classification_report(y_test, y_pred))
    
    # Store results
    results[model_name] = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred)
    }
```

```python
# Prepare data for plotting
plot_data = []
for model_name, metrics_dict in results.items():
    for metric, value in metrics_dict.items():
        plot_data.append({'Model': model_name, 'Metric': metric, 'Value': value})

# Convert to DataFrame
plot_df = pd.DataFrame(plot_data)

# Create the bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Metric', y='Value', hue='Model', data=plot_df)
plt.title('Model Performance Comparison')
plt.ylabel('Score')
plt.ylim(0, 1)
plt.show()
```

## 4: Model Development on Redued Data (after correlation)

```python
# Set the correlation threshold
correlation_threshold = 0.9

# Create a correlation matrix
corr_matrix = raw_df_scaled.corr()

# Identify columns to drop
columns_to_drop = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > correlation_threshold:
            colname = corr_matrix.columns[i]
            columns_to_drop.add(colname)

# Drop the identified columns
raw_df_reduced = raw_df_scaled.drop(columns=columns_to_drop)

# Display the columns to be dropped
print("Columns to be dropped due to high correlation:")
print(columns_to_drop)

# Display the reduced dataframe
raw_df_reduced.head()
```

```python
# Define the features (X) and target (y)
X = raw_df_reduced.drop(columns=['class1', 'class1_binary'])
y = raw_df_reduced['class1_binary']

# Split the dataset into training and testing sets (70:30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    'SVM': SVC()
}


# Train and evaluate models
results = {}
for model_name, model in models.items():
    # Cross-validation
    cv_scores = cross_val_score(model, X_train, y_train, cv=5)
    print(f"{model_name} CV Accuracy: {cv_scores.mean():.4f} Â± {cv_scores.std():.4f}")
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on test data
    y_pred = model.predict(X_test)
    
    # Classification report
    print(f"\n{model_name} Classification Report:\n")
    print(classification_report(y_test, y_pred))
    
    # Store results
    results[model_name] = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred)
    }


# Prepare data for plotting
plot_data = []
for model_name, metrics_dict in results.items():
    for metric, value in metrics_dict.items():
        plot_data.append({'Model': model_name, 'Metric': metric, 'Value': value})

# Convert to DataFrame
plot_df = pd.DataFrame(plot_data)

# Create the bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Metric', y='Value', hue='Model', data=plot_df)
plt.title('Model Performance Comparison')
plt.ylabel('Score')
plt.ylim(0, 1)
plt.show()
```

Insight:

## 5: Final model using Random Forest with Hyper parameter optimization

```python
from sklearn.model_selection import GridSearchCV
import seaborn as sns

import matplotlib.pyplot as plt

# Define the features (X) and target (y)
X = raw_df_reduced.drop(columns=['class1', 'class1_binary'])
y = raw_df_reduced['class1_binary']

# Split the dataset into training and testing sets (70:30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the parameter grid for hyperparameter optimization
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Initialize the Random Forest model
rf = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV with 5-fold cross-validation
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=1, verbose=1, scoring='recall') # Recall since we want to focus specially on deteccting all VPN apps

# Fit the model
grid_search.fit(X_train, y_train)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Cross-Validation Score:", best_score)

# Train the final model with the best parameters
final_rf = RandomForestClassifier(**best_params, random_state=42)
final_rf.fit(X_train, y_train)

# Predict on the test data
y_pred = final_rf.predict(X_test)

# Classification report
print("\nFinal Random Forest Classification Report:\n")
print(classification_report(y_test, y_pred))

# Cross-validation scores for the final model
cv_scores = cross_val_score(final_rf, X_train, y_train, cv=5)

# Prepare data for plotting
plot_data = [{'Metric': 'Accuracy', 'Value': cv_scores.mean()}]

# Convert to DataFrame
plot_df = pd.DataFrame(plot_data)

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Metric', y='Value', data=plot_df)
plt.title('Cross-Validation Scores for Final Random Forest Model')
plt.ylabel('Score')
plt.ylim(0, 1)
plt.show()
```

```python
# Get feature importances from the final Random Forest model
importances = final_rf.feature_importances_

# Create a DataFrame for better visualization
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
})

# Sort the DataFrame by importance
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Plot the feature importances
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance from Final Random Forest Model')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

## 6: Conclusion:

The top 5 important features of determining the VPN vs. Non-VPN traffic are:


#### **1. total_biat (Total Backward Inter-Arrival Time)**  
This feature measures the **total time between packets traveling backward** in a communication flow.  

- **VPN Traffic:** Often shows more **uniform or delayed backward inter-arrival times** due to packet encapsulation and encryption overhead.  
- **Non-VPN Traffic:** Typically has **more natural variations** in backward timings since no encryption or additional layers alter packet behavior.  

**Insight:** *If backward packet timing is consistent and slightly delayed, itâ€™s more likely VPN traffic.*

---

#### **2. min_flowiat (Minimum Flow Inter-Arrival Time)**  
This captures the **shortest time gap between two packets in the flow**.  

- **VPN Traffic:** Encryption and encapsulation often cause **slight delays**, increasing the minimum inter-arrival time.  
- **Non-VPN Traffic:** Direct communication typically allows **faster inter-arrival times** with lower delays.  

**Insight:** *A slightly higher minimum inter-arrival time might hint at VPN encapsulation overhead.*

---

#### **3. flowBytesPerSecond (Number of Flow Bytes per Second)**  
This feature measures the **rate of data transmission per second** in a flow.  

- **VPN Traffic:** Tends to have **more consistent or throttled data flow rates** because VPN protocols (like OpenVPN or WireGuard) standardize packet sizes and flow rates to reduce detectability.  
- **Non-VPN Traffic:** Displays **variable data flow rates** based on application-specific behaviors (e.g., streaming vs. text communication).  

**Insight:** *A steady and regulated byte-per-second rate often suggests VPN traffic.*

---

#### **4. total_fiat (Total Forward Inter-Arrival Time)**  
This measures the **total time between packets traveling forward** in the communication flow.  

- **VPN Traffic:** Forward inter-arrival times might exhibit **more uniformity** due to VPN buffering mechanisms.  
- **Non-VPN Traffic:** More **variability** is expected as different applications and protocols send packets based on real-time demands.  

**Insight:** *If total forward inter-arrival time shows uniform patterns, it leans towards VPN traffic.*

---

#### **5. max_flowiat (Maximum Flow Inter-Arrival Time)**  
This captures the **longest time gap between two packets in the flow**.  

- **VPN Traffic:** Long inter-packet delays might be **less frequent** due to buffering mechanisms in VPN tunnels.  
- **Non-VPN Traffic:** Natural variability might result in **occasional large gaps** between packets.  

**Insight:** *Consistent maximum flow inter-arrival times are a subtle indicator of VPN traffic.*
